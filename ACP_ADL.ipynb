```python
import numpy as np
import sklearn.preprocessing
```

```python
#QUESTION 01


X = np.array([[1, -1, 2],
              [2, 0, 0],
              [0, 1, -1]])
```

```python
#QUESTION 02


# Affichage de la matrice X
print("Matrice X :")
print(X)

# Calcul de la moyenne de X
moyenne_X = np.mean(X)
print("\nMoyenne de X :", moyenne_X)

# Calcul de la variance de X
variance_X = np.var(X)
print("Variance de X :", variance_X)
```

```python
#QUESTION 03

from sklearn.preprocessing import scale
# Normalisation de la matrice X
X_normalized = scale(X)

# Affichage de la matrice normalisée
print("Matrice X normalisée :")
print(X_normalized)
```

On constate que nous avons fait une normalisation de type Z-score, donc nos valeurs sont toute centrées autour de la moyenne 0.

```python
#QUESTION 04

# Calcul de la moyenne de la matrice normalisée
moyenne_X_normalized = np.mean(X_normalized)
print("Moyenne de la matrice normalisée :", moyenne_X_normalized)

# Calcul de la variance de la matrice normalisée
variance_X_normalized = np.var(X_normalized)
print("Variance de la matrice normalisée :", variance_X_normalized)
```

La moyenne de la matrice normalisée sera très proche de zéro (généralement avec une précision numérique), car la normalisation a pour effet de centrer les données autour de zéro en soustrayant la moyenne de chaque colonne à chaque valeur de la colonne. Cela garantit que la moyenne de chaque colonne de la matrice normalisée est approximativement égale à zéro.

La variance de la matrice normalisée sera approximativement égale à 1, car la normalisation met également les données à l'échelle de manière à avoir un écart-type (la racine carrée de la variance) égal à 1. Cela signifie que la dispersion des valeurs dans chaque colonne est mise à l'échelle de manière à avoir une variance unitaire. La normalisation est utile pour s'assurer que toutes les caractéristiques ont la même importance lors de l'analyse des données ou de l'entraînement de modèles d'apprentissage automatique, car elle élimine l'impact des échelles différentes des caractéristiques sur les résultats.

```python
#QUESTION 02


moyenne_variables = np.mean(X2, axis=0)

print("Moyenne sur les variables :", moyenne_variables)
```

```python
#QUESIONT 03

from sklearn.preprocessing import MinMaxScaler

# Création de l'objet MinMaxScaler
scaler = MinMaxScaler()

# Normalisation de la matrice X dans l'intervalle [0, 1]
X2_normalized = scaler.fit_transform(X2)

# Affichage de la matrice normalisée
print("Matrice X2 normalisée dans [0, 1] :")
print(X2_normalized)

# Calcul de la moyenne des variables
moyenne_variables = np.mean(X2_normalized, axis=0)
print("Moyenne des variables normalisées :", moyenne_variables)
```

les valeurs de la matrice normalisée sont désormais toutes comprises entre 0 et 1, comme spécifié.
La normalisation Min-Max met à l'échelle les données de manière à ce que la valeur minimale devienne 0 et la valeur maximale devienne 1, tout en maintenant la relation proportionnelle entre les valeurs.
La moyenne des variables normalisées dépendra des valeurs initiales de la matrice X, mais elles sont proche de 0.5

```python
#QUESTION 01

from sklearn.datasets import load_iris

# Chargement des données Iris
iris = load_iris()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Charger les données Iris
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# Liste des combinaisons de variables (2 à 2)
combinations = [(0, 1), (0, 2), (0, 3), (1, 2), (1, 3), (2, 3)]
plot_colors = ['r', 'g', 'b']
plot_markers = ['o', 's', '^']

# Créez 2x3 pour les six combinaisons de variables
plt.figure(figsize=(12, 8))
for i, (x1, x2) in enumerate(combinations):
    plt.subplot(2, 3, i + 1)
    for target, color, marker in zip(range(len(target_names)), plot_colors, plot_markers):
        plt.scatter(X[y == target, x1], X[y == target, x2], color=color, marker=marker, label=target_names[target])
    plt.xlabel(iris.feature_names[x1])
    plt.ylabel(iris.feature_names[x2])
    plt.legend(loc='best')
    plt.title(f'{iris.feature_names[x1]} vs {iris.feature_names[x2]}')

plt.tight_layout()
plt.show()
```

petal length vs petal width semble être la meilleure visualisation pour les données Iris, car elle permet de distinguer clairement les trois classes. Les setosas sont généralement bien séparés des versicolors et des virginicas sur cette projection.

```python
#QUESTION 01


from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
#QUESTION 02


# Créez des objets PCA et LDA
pca = PCA(n_components=2)
lda = LDA(n_components=2)
# Appliquez PCA et LDA aux données Iris
irisPCA = pca.fit(X).transform(X)
irisLDA = lda.fit(X, y).transform(X)
```

```python
#QUESTION 03



# Créez un nuage de points pour PCA
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(irisPCA[:, 0], irisPCA[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.title('ACP')
plt.xlabel('Premier Composant Principal')
plt.ylabel('Deuxième Composant Principal')

# Créez un nuage de points pour LDA
plt.subplot(1, 2, 2)
plt.scatter(irisLDA[:, 0], irisPCA[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.title('ADL')
plt.xlabel('Premier Composant Discriminant')
plt.ylabel('Deuxième Composant Discriminant')

plt.show()
```

Dans cette visualisation, chaque point est coloré en fonction de sa classe d'iris (setosa, versicolor, virginica).

La différence principale entre les deux visualisations est que l'ACP vise à maximiser la variance totale des données dans les nouvelles dimensions, tandis que l'ADL cherche à maximiser la variance entre les classes tout en minimisant la variance à l'intérieur des classes. En conséquence, l'ADL tend à séparer davantage les classes les unes des autres, ce qui est clairement visible dans la visualisation ADL.

En résumé, l'ADL est une technique de réduction de dimensionnement supervisée qui tient compte des classes, tandis que l'ACP est une technique non supervisée qui ne tient pas compte des classes, ce qui explique la différence dans les visualisations.
