```python
from sklearn.datasets import make_blobs

x, y = make_blobs(n_samples=1000, n_features=2, centers=4)
x.size
```

```python
import matplotlib.pyplot as plt

# Génère 1000 données réparties en 4 clusters avec 2 variables
X, y = make_blobs(n_samples=1000, n_features=2, centers=4,random_state = 3)

# Affiche le nuage de points des données générées
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('Nuage de points avec 4 clusters')
plt.xlim(-15,15)
plt.ylim(-15,15)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Génère 1000 données réparties en 4 clusters avec 2 variables
x1, y1 = make_blobs(n_samples=100, n_features=2, centers=2)
x2, y2 = make_blobs(n_samples=500, n_features=2, centers=3)
X = np.vstack((x1,x2))
y = np.hstack((y1,y2))
print(X.shape)
# Affiche le nuage de points des données générées
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('Nuage de points')
plt.xlim(-15,15)
plt.ylim(-15,15)
plt.show()
```
